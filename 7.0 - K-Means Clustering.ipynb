```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_regression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from matplotlib.patches import Circle
```

```python
np.random.seed(42)
class_1 = np.random.normal(loc=[6, 1], scale=1, size=(60, 2))
class_2 = np.random.normal(loc=[1, 6], scale=1, size=(80, 2))
class_3 = np.random.normal(loc=[-1, -1], scale=1, size=(50, 2))

X = np.vstack([class_1, class_2, class_3])
# y = np.hstack([np.ones(100), np.zeros(100)])

plt.scatter(X[:, 0], X[:, 1], marker='o')
plt.title('Generated Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
n_clusters = 3
n_iters = 20
k = 3

centroids = X[np.random.choice(len(X), k, replace=False)]
```

```python
for _ in range(n_iters):
    
    X_expanded = X[:, np.newaxis]
    
    distances = np.linalg.norm(X_expanded - centroids, axis=2)
    
    labels = np.argmin(distances, axis=1)

    new_centroids = np.array([X[labels == i].mean(axis=0) for i in range(k)])

    if np.all(centroids == new_centroids):
        break
    
    centroids = new_centroids
```

```python
print("Cluster labels:\n", labels)
print("Final centroids:\n", centroids)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap="viridis", marker='o')
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=100, color='red', label='Centroids')
plt.title('Clustered Datapoints')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
# plt.legend()
plt.show()
```
