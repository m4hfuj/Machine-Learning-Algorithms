```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_regression
```

```python
data = pd.read_csv("Datasets/score/score_updated.csv")
data.head()
```

```python
X = data.iloc[:, 0].to_numpy().reshape(-1, 1)
y = data.iloc[:, 1].to_numpy().reshape(-1, 1)
```

```python
sns.scatterplot(x=X.squeeze(), y=y.squeeze())
plt.title('Regression Dataset')
plt.show()
```

```python
def cost_function(y, y_hat):
    return ((y - y_hat)**2).mean()
```

```python
n_epochs = 200

m = len(y)
alpha = 0.01
w1 = 1
w0 = 1

for epoch in range(n_epochs):
    y_hat = w0 + w1*X
    cost = cost_function(y, y_hat)

    # dw = 1/m * summation(y_hat - y)*X
    dw1 = alpha * ((y_hat - y) * X).mean()
    # db = 
    dw0 = alpha * (y_hat - y).mean() 
    
    w1 -= dw1
    w0 -= dw0
```

```python
y_pred = w0 + w1*X
```

```python
print("w0  :", w0)
print("w1  :", w1)
print("Cost:", cost_function(y, y_pred))
```

```python
sns.scatterplot(x=X.squeeze(), y=y.squeeze())
plt.plot([X.max(), X.min()], [y_pred.max(), y_pred.min()], color='r')
plt.title('Regression Dataset')
plt.show()
```
